## **1.3 e 1.4 Implementação do Algoritmo em Python e Análise Energética**


Nesta etapa, foi implementado em Python o algoritmo de verificação de segurança da Nave Orion.

O script simula a leitura dos dados da telemetria, realiza as validações com base nas faixas de segurança predefinidas e, ao final, informa se a nave está “PRONTA PARA DECOLAR” ou se a “DECOLAGEM FOI ABORTADA”.

Além disso, o programa realiza o cálculo da autonomia energética considerando capacidade das baterias, carga atual, consumo estimado na decolagem e perdas energéticas.

<br>

##**Detalhamento do Cálculo Energético**

<br>

Com base nos dados do cenário aprovado:

Capacidade total: 13 kWh

Carga atual: 95%

Perdas energéticas: 8%

Consumo estimado: 8,5 kWh/h

Energia disponível:

13 × 0,95 = 12,35 kWh

Energia útil após perdas:

12,35 × 0,92 = 11,36 kWh

Autonomia:

11,36 ÷ 8,5 ≈ 1,33 horas

**Convertendo:
≈ 1h20min de autonomia**



```python
def verificacao():

    print("Bem vindos ao algoritmo de segurança da Nave Orion!")

    nome = input("Qual é o seu nome astronauta? ")

    saudacao = input(f"{nome}, está preparado para a verificação de segurança para a decolagem? S/N ").strip().lower() #.strip() serve para tirar os espaçamentos que pode haver na frente e de trás da resposta do usuário, evitando assim um erro

    if saudacao == "sim" or saudacao == "s":
        print("Iniciando verificação de segurança...")
    else:
        print("Abortando missão!")
        return

    temperatura_interna = float(input("Digite a temperatura interna da cabine (°C): "))

    if temperatura_interna >= 18 and temperatura_interna <= 26:
        print(f"A Temperatura interna está em °C {temperatura_interna}, está dentro dos parâmetros...")
    else:
        print(f"A Temperatura interna está em °C {temperatura_interna}, está fora dos parâmetros, abortando decolagem...")
        return

    temperatura_externa = int(input("Digite a temperatura externa da nave (°C): "))

    if temperatura_externa >= 0 and temperatura_externa <= 2760:
        print(f"A Temperatura externa está em °C {temperatura_externa}, está dentro dos parâmetros...")
    else:
        print(f"A Temperatura externa está em °C {temperatura_externa}, está fora dos parâmetros, abortando decolagem...")
        return

    tanques_propelente = float(input("Digite a pressão nos tanques de propelente: "))

    if tanques_propelente >= 20 and tanques_propelente <= 30:
        print(f"A Pressão dos tanques propelente estão em {tanques_propelente} bars, está dentro dos parâmetros...")
    else:
        print(f"A Pressão dos tanques propelente estão em {tanques_propelente} bars, está fora dos parâmetros, abortando decolagem...")
        return

    tanque_gases = float(input("Digite a pressão dos tanques de gás (Hélio / O² / N²): "))

    if tanque_gases >= 350 and tanque_gases <= 420:
        print(f"A Pressão dos tanques de gás estão em {tanque_gases} bars, está dentro dos parâmetros...")
    else:
        print(f"A Pressão dos tanques de gás estão em {tanque_gases} bars, está fora dos parâmetros, abortando decolagem...")
        return

    energia_solar = float(input("Digite quantos kWs estão sendo gerados pela energia solar: "))

    if energia_solar >= 9 and energia_solar <= 11.2:
        print(f"A quantidade de energia solar gerada é de {energia_solar} kWs, está dentro dos parâmetros...")
    else:
        print(f"A quantidade de energia solar gerada é de {energia_solar} kWs, está fora dos parâmetros, abortando decolagem...")
        return

    armazenamento_bateria = float(input("Digite a capacidade de armazenamento das baterias (kWh): "))

    if armazenamento_bateria >= 12 and armazenamento_bateria <= 14.4:
        print(f"A quantidade de armazenamento das baterias é de {armazenamento_bateria} kWh, está dentro dos parâmetros...")
    else:
        print(f"A quantidade de armazenamento das baterias é de {armazenamento_bateria} kWh, está fora dos parâmetros, abortando decolagem...")
        return

    carga_bateria = float(input("Digite a porcentagem de carga da bateria (%): "))

    if carga_bateria >= 94 and carga_bateria <= 97:
        print(f"A porcentagem de carga na bateria é de % {carga_bateria}, está dentro dos parâmetros...")
    else:
        print(f"A porcentagem de carga na bateria é de % {carga_bateria}, está fora dos parâmetros, abortando decolagem...")
        return

    print("Agora iremos iniciar a verificação estrutural da nave, utilize '0' para falha e '1' para OK...")

    cm = int(input("Módulo de Tripulação (CM) está operacional? "))

    if cm == 1:
        print("Módulo de Tripulação (CM) OK...")
    else:
        print("Módulo de Tripulação (CM) apresentou falha, abortando decolagem")
        return

    esm = int(input("Módulo de serviço (ESM) está operacional? "))

    if esm == 1:
        print("Módulo de serviço (ESM) OK...")
    else:
        print("Módulo de serviço (ESM) apresentou falha, abortando decolagem")
        return

    las = int(input("Sistema de Aborto (LAS) está operacional? "))

    if las == 1:
        print("Sistema de Aborto (LAS) OK...")
    else:
        print("Sistema de Aborto (LAS) apresentou falha, abortando decolagem")
        return

    osa = int(input("Adaptador de Estágio (OSA) está operacional? "))

    if osa == 1:
        print("Adaptador de Estágio (OSA) OK...")
    else:
        print("Adaptador de Estágio (OSA) apresentou falha, abortando decolagem")
        return

    integridade_geral = input("Tem certeza que está tudo OK com a integridade geral da nave? S/N ").strip().lower()

    if integridade_geral == "sim" or integridade_geral == "s":
        print("Verificação estrutural finalizada, nave liberada para decolagem!")
    else:
        print("Integridade geral negada, abortando decolagem...")
        return

    print("Começando cálculo da autonomia energética...")

    perdas_energeticas = 0.08 # a nave perde 8% ao decolar
    energia_disponivel = armazenamento_bateria * (carga_bateria / 100) # o /100 serve para transformar o número em decimal antes de multiplicar a var armazenamento_bateria pelo resultado
    energia_util = energia_disponivel * (1 - perdas_energeticas) # o 1 Simboliza 100% o - perdas_energeticas 0.08, logo virara 0.92
    consumo_decolagem = 8.5 # var que guarda o valor fixo da decolagem por hora

    autonomia = energia_util / consumo_decolagem # faz a divisão da energia util e consumo decolagem, o valor vira a autonomia da nave

    horas = int(autonomia) # transforma o número decimal (autonomia) em int
    minutos = (autonomia - horas) * 60 # separamos a parte decimal ( pré minutos ) da parte int (horas) e fazemos *60 p dar os minutos

    print(f"Com os dados fornecidos, a nave tem {horas}h e {minutos:.0f}min de autonomia energética!") # :.0f depois de mostrar o valor da var minutos, é para remover as casas decimais do que vieram da última etapa

verificacao()
```

> ## **Ética e responsabilidade**

<br>



A introdução da Inteligência Artificial em diversas esferas da sociedade trouxe consigo avanços significativos, mas também levantou questões éticas urgentes. Quando desenvolvida ou utilizada sem diretrizes claras, a IA pode perpetuar preconceitos e discriminações já existentes na sociedade. Buolamwini e Gebru (2018) demonstraram, por exemplo, que sistemas de reconhecimento facial apresentam taxas de erro significativamente maiores ao identificar mulheres e pessoas negras, evidenciando que algoritmos não são neutros, pois são moldados pelos dados com os quais são treinados.

A falta de transparência nas decisões automatizadas representa outro grande desafio ético. Em áreas sensíveis como saúde, crédito financeiro e segurança, os cidadãos têm o direito de compreender como as decisões que afetam suas vidas são tomadas. Sem essa explicabilidade, o risco de injustiças sistemáticas aumenta consideravelmente, comprometendo direitos fundamentais.

No campo do trabalho, Brynjolfsson e McAfee (2014) alertam que a automação pode redefinir profundamente o mercado, substituindo trabalhadores em tarefas repetitivas e de baixa qualificação. Embora o aumento de produtividade seja inegável, cabe ao Estado, às empresas e à sociedade civil garantir, em conjunto, políticas e práticas que permitam à sociedade se adaptar a essa nova realidade com equidade.

Por fim, a responsabilidade ética não deve ser uma reflexão posterior ao desenvolvimento tecnológico, mas sim incorporada desde o início , o conceito conhecido como **ética by design**. Essa abordagem propõe que preocupações éticas, como privacidade, equidade e transparência, sejam consideradas já na fase de concepção e projeto da tecnologia, e não apenas após sua implementação. Isso implica curadoria dos dados utilizados no treinamento, mitigação ativa de vieses e criação de controles que garantam que a IA sirva ao bem-estar humano e não o comprometa.

> ## **Sustentabilidade tecnológica**

<br>

A tecnologia e a sustentabilidade são dimensões cada vez mais inseparáveis no contexto do desenvolvimento contemporâneo. À medida que a transformação digital avança, torna-se imprescindível que a inovação tecnológica seja orientada não apenas pela busca de eficiência e lucro, mas também pela responsabilidade ambiental e social. Crescer sem considerar os impactos gerados deixou de ser uma opção aceitável e os profissionais de tecnologia estão no centro dessa mudança.

A transformação digital tem impulsionado o desenvolvimento econômico em todo o planeta, mas agora é necessário refinar esse olhar e entender a tecnologia também como uma ferramenta de proteção ambiental. Conforme aponta a Ascenty (2024), a sustentabilidade exige que a inovação tecnológica leve em consideração os impactos ambientais e sociais de curto e longo prazo, permitindo que empresas otimizem recursos, reduzam emissões e adotem modelos de negócios mais responsáveis.

Na prática, esse movimento se traduz em diversas frentes. Na eficiência energética, soluções como iluminação LED, sensores inteligentes e energia solar já fazem parte da realidade de empresas e cidades. Na gestão de recursos naturais, o reaproveitamento de água, a agricultura de precisão e a dessalinização representam respostas concretas à escassez hídrica. No campo da mobilidade, carros elétricos, transporte público limpo e logística reversa reduzem significativamente as emissões de carbono. Por fim, práticas como compostagem, reaproveitamento de materiais e impressão 3D consolidam o conceito de economia circular, onde o resíduo de um processo se torna insumo de outro.

Segundo o SIDI (2024), empresas devem alinhar-se a frameworks reconhecidos, como os Objetivos de Desenvolvimento Sustentável da ONU, integrando práticas com legitimidade e responsabilidade. O futuro da tecnologia não pode ser dissociado do futuro do planeta, e cabe aos profissionais de tecnologia assumir esse papel de protagonismo na construção de um desenvolvimento verdadeiramente sustentável.
